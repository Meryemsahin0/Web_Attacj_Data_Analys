```python
import pandas as pd

data = {
    "Country": ["US", "UK", "CA", "AU", "DE"],
    "Total Library Size": [1000, 1500, 800, 1200, 2000],
    "No. of TV Shows": [500, 600, 400, 700, 800],
    "No. of Movies": [500, 900, 400, 500, 1200],
    "Cost Per Month - Basic": [10, 8, 12, 10, 15],
    "Cost Per Month - Standard": [15, 12, 18, 15, 20],
    "Cost Per Month - Premium": [20, 15, 25, 20, 30]
}

df = pd.DataFrame(data)


print(df.info())


for column in df.columns:
    unique_values = df[column].unique()
    print(f"Column: {column}, Unique Values: {unique_values}")


numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
categorical_columns = df.select_dtypes(include=['object']).columns

print("Numeric Columns:", numeric_columns)
print("Categorical Columns:", categorical_columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


def plot_categorical_variables(df):
    categorical_columns = df.select_dtypes(include=['object']).columns
    for column in categorical_columns:
        plt.figure(figsize=(8, 6))
        sns.countplot(data=df, x=column)
        plt.title(f"{column} Distribution")
        plt.xlabel(column)
        plt.ylabel("Count")
        plt.xticks(rotation=45)
        plt.show()


def plot_numerical_variables(df):
    numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
    for column in numeric_columns:
        plt.figure(figsize=(8, 6))
        sns.histplot(data=df, x=column, kde=True)
        plt.title(f"{column} Distribution")
        plt.xlabel(column)
        plt.ylabel("Frequency")
        plt.show()


data = {
    "Country": ["US", "UK", "CA", "AU", "DE"],
    "Total Library Size": [1000, 1500, 800, 1200, 2000],
    "No. of TV Shows": [500, 600, 400, 700, 800],
    "No. of Movies": [500, 900, 400, 500, 1200],
    "Cost Per Month - Basic": [10, 8, 12, 10, 15],
    "Cost Per Month - Standard": [15, 12, 18, 15, 20],
    "Cost Per Month - Premium": [20, 15, 25, 20, 30]
}

df = pd.DataFrame(data)


plot_categorical_variables(df)


plot_numerical_variables(df)
```

```python


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = {
    "Country": ["US", "UK", "CA", "AU", "DE"],
    "Total Library Size": [1000, 1500, 800, 1200, 2000],
    "No. of TV Shows": [500, 600, 400, 700, 800],
    "No. of Movies": [500, 900, 400, 500, 1200],
    "Cost Per Month - Basic": [10, 8, 12, 10, 15],
    "Cost Per Month - Standard": [15, 12, 18, 15, 20],
    "Cost Per Month - Premium": [20, 15, 25, 20, 30]
}

df = pd.DataFrame(data)


sns.pairplot(df[['Total Library Size', 'No. of TV Shows', 'No. of Movies']])
plt.show()


df_cost = df[['Cost Per Month - Basic', 'Cost Per Month - Standard', 'Cost Per Month - Premium']]
sns.boxplot(data=df_cost)
plt.show()
import plotly.express as px

fig = px.bar(df, x='Country', y='Total Library Size', title='Total Library Size by Country')
fig.show()


fig = px.scatter(df, x='Total Library Size', y='No. of Movies', color='Country', title='Total Library Size vs. No. of Movies')
fig.show()
```
